```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
energy_data = pd.read_csv("C:/Users/HP/Downloads/impacts-of-energy-sources.csv")
energy_data.sample(10)
```

```python
energy_data.shape
energy_data[energy_data.isnull().any(axis=1)]
```

```python
energy_data.isnull().sum()
```

```python
energy_data['Total land use'] = energy_data['Total land use'].fillna(energy_data['Total land use'].mean())
energy_data.isnull().sum()
```

```python
energy_data.head()
```

```python
energy_data['log_greenhouse_gas_emissions'] = np.log(energy_data['Greenhouse gas emissions'] + 1)
energy_data.head()
```

```python
features = ['Freshwater eutrophication', 'Water use', 'Non-carcinogenic toxicity',
             'Molybdenum', 'Nickel',
            'Uranium']

label = energy_data['log_greenhouse_gas_emissions']
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6, 6))
ax[0].hist(label, bins=100)
ax[0].set_ylabel('Frequency')

ax[0].axvline(x=label.mean(), color='cyan', linestyle='dashed', linewidth=1)
ax[0].axvline(x=label.median(), color='magenta', linestyle='dashed', linewidth=1)
ax[0].axvline(x=label.mode()[0], color='grey', linestyle='dashed', linewidth=1)

ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Values')
plt.suptitle('label distribution')
plt.show()

```

```python
for col in features:
    ax = fig.gca()
    features = energy_data[col]
    correlation = features.corr(label)
    plt.scatter(x=features, y=label)
    plt.xlabel(col)
    plt.ylabel('Green house')
    plt.title('Correlation ' + col + str(correlation))
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = energy_data[['Freshwater eutrophication', 'Water use', 'Non-carcinogenic toxicity','Molybdenum', 'Nickel','Uranium']].values
y = energy_data[['log_greenhouse_gas_emissions']].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
y = np.ravel(y)
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
model = GradientBoostingRegressor()
model.fit(x_train, y_train)

pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, pred)
print('MSE: ', mse)

rmse = np.sqrt(mse)
print('RMSE: ', rmse)

R2 = r2_score(y_test, pred)
print("R2: ", R2)

print(y_test.shape)
print(pred.shape)

plt.scatter(y_test, pred)
plt.xlabel('Actual labels')
plt.ylabel('Predicted labels')


z = np.polyfit(y_test.flatten(), pred, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test), color='magenta')
plt.show()
```
